# Bài 1 - Vẽ phương trình y = ax + b

```python
import matplotlib.pyplot as plt
```

```python
def ve_pt_bac_1(a,b):
    #Check a # 0 
    if a != 0:
        fig = plt.figure()
        ax = fig.add_subplot(1,1,1)
        ax.set_axis_off
        ax.plot([0,-b/a],[b,0])
        
    else: print("Please enter a # 0")
ve_pt_bac_1(-2,5)
```

# Bài 2 - Đọc và vẽ dữ liệu chứng khoán

```python
# Đọc dữ liệu
import csv
import matplotlib
from matplotlib import pyplot as plt

close_price = []
maCK = "vjc"

with open(f"./files/chungkhoan/excel_{maCK}.csv", mode="r", encoding="utf-8") as f:
    stock_data = csv.reader(f)
    for row in stock_data:
           close_price.append(row[10])           
# Xoá header
del(close_price[0])

#Convert to float
close_price_converted = []
for price in close_price:
    close_price_converted.append(float(price))
    
close_price_converted

# vẽ biểu đồ

plt.figure(figsize=(14,8)).patch.set_facecolor('#ad5e58')
plt.plot(close_price_converted, color="#ad5e58", label=f"{maCK.upper()} close price")
plt.title(f"Giá đóng cửa của {maCK.upper()}", fontsize = 20)
plt.xlabel("Total Number", fontsize = 16)
plt.ylabel("Price",fontsize = 16)
plt.legend(loc=[1.1,0.5])
plt.show()
```

# Đọc nhiều mã chứng khoán khác nhau

```python
# Đọc dữ liệu
import csv
import random
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.colors as colors

#Get colors list from matplotlib
colors = list(colors._colors_full_map.values())

def GiaCK(maCK):
    close_price = []   
   
    with open(f"./files/chungkhoan/excel_{maCK}.csv", mode="r", encoding="utf-8") as f:
        stock_data = csv.reader(f)
        for row in stock_data:
               close_price.append(row[10])           
    # Xoá header
    del(close_price[0])

    #Convert to float
    close_price_converted = []
    for price in close_price:
        close_price_converted.append(float(price))    
   
    # vẽ biểu đồ
    plt.figure(figsize=(14,8)).patch.set_facecolor(colors[random.randint(0,len(colors))])
    plt.plot(close_price_converted, color=colors[random.randint(0,len(colors))], label=f"{maCK.upper()} close price")
    plt.title(f"Giá đóng cửa của {maCK.upper()}", fontsize = 20, color='w')
    plt.xlabel("Số lần giao dịch", fontsize = 16,color='w')
    plt.ylabel("Price",fontsize = 16,color='w')
    plt.xticks(color = "w")
    plt.legend(loc=[1.1,0.5])
    plt.show()
# Goi ham
GiaCK("vnm")
```
