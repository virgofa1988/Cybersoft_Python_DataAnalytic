# Bài 2 - Đọc và vẽ dữ liệu chứng khoán

```python
# Đọc dữ liệu
import csv
import matplotlib
from matplotlib import pyplot as plt

close_price = []
maCK = "vnm"

with open(f"./files/chungkhoan/excel_{maCK}.csv", mode="r", encoding="utf-8") as f:
    stock_data = csv.reader(f)
    for index,row in enumerate(stock_data):
            # Bỏ đi header
            if(index == 0):continue
            close_price.append(float(row[10]))   

# vẽ biểu đồ

plt.figure(figsize=(14,8)).patch.set_facecolor('#ad5e58')
plt.plot(close_price, color="#ad5e58", label=f"{maCK.upper()} close price")
plt.title(f"Giá đóng cửa của {maCK.upper()}", fontsize = 20)
plt.xlabel("Total Number", fontsize = 16)
plt.ylabel("Price",fontsize = 16)
plt.legend(loc=[1.1,0.5])
plt.show()
```

# Đọc nhiều mã chứng khoán khác nhau

```python
# Đọc dữ liệu
import csv
import random
import matplotlib
from matplotlib import pyplot as plt
import matplotlib.colors as colors

#Get colors list from matplotlib
colors = list(colors._colors_full_map.values())

#Tạo hàm để truyền vào mã CK muốn xem biểu đồ giá
def GiaCK(maCK):
    close_price = []   
   
    with open(f"./files/chungkhoan/excel_{maCK}.csv", mode="r", encoding="utf-8") as f:
        stock_data = csv.reader(f)
        for index,row in enumerate(stock_data):
            # Bỏ đi header
            if(index == 0):continue
            close_price.append(float(row[10]))                  
         
    # vẽ biểu đồ
    plt.figure(figsize=(14,8)).patch.set_facecolor(colors[random.randint(0,len(colors))])
    plt.plot(close_price, color=colors[random.randint(0,len(colors))], label=f"{maCK.upper()} close price")
    plt.title(f"Giá đóng cửa của {maCK.upper()}", fontsize = 20, color='w')
    plt.xlabel("Số lần giao dịch", fontsize = 16,color='w')
    plt.ylabel("Price",fontsize = 16,color='w')
    plt.xticks(color = "w")
    plt.legend(loc=[1.1,0.5])
    plt.show()
# Goi ham
GiaCK("flc")
```

# Bài 3 - Vẽ phương trình bậc 2: y=ax^2 + bx + c (-5 <= x <=5)

```python
import math
import numpy as np
def PTBac_2(a,b,c):
   # Kiểm tra các hệ số
    if (a == 0):
        if(b == 0): print('Phương trình vô nghiệm !')
        else: 
            print(f'Phương trình bậc 1 có 1 nghiệm x = {-c/b}')
            plt.plot([0,-c/b],[c,0])
            plt.show()
    # Tính nghiệm lúc a !=0
    delta = b * b - 4*a*c
    # Tính Nghiệm
    if(delta > 0):
        # Giá trị x trong khoảng -5 đến 5
        x = np.array(np.arange(-5,5,0.1))
        y = a*x*x + b*x + c        
        plt.plot(x,y,label=f'y ={a}*x^2 + {b}*x + {c}')
        plt.legend()
        plt.grid(alpha=0.4)
        plt.show()
    elif (delta == 0):
        print('Phương trình có 1 nghiệm kép')
        x1 = (-b / (2*a));
        plt.plot([0,x1],[c,0])
    else:
        print('Phương trình vô nghiệm !')
```

```python
PTBac_2(10,9,1)
```

# Bài tập về supermarket

```python
import matplotlib
from matplotlib import pyplot as plt
import random
import csv

columns = []
data = []
# Đọc dữ liệu từ file và tách columns và dữ liệu riêng
with open('./files/supermarket_sale_vn/supermarket_sales_vn.csv', mode ='r',encoding='utf-8') as f:
    data_supermarket = csv.DictReader(f,delimiter=',')
    
    #get full data
    for index, row in enumerate(data_supermarket):
        data.append(row)
    
    #Get all keys
    for index, row in enumerate(data_supermarket):
        columns = list(row.keys())
        if index ==0: break
```

```python
# Tách city list và gender list
city_list = []
city_name =[]
gender_list = []
gender_name=[]
for row in data:
    city_list.append(row['City'])
    city_name = set(city_list)
    gender_list.append(row['Gender'])
    gender_name = set(gender_list)
    
```

Vẽ biểu đồ cột cho tần suất, tần số đơn hàng theo địa phương

```python
# Tạo biến chưa số liệu theo city name và vẽ biểu đồ cột
data_city = dict.fromkeys(city_name,0)

for city in data_city:
    for row in data:
        if(city == row['City']): data_city[city] +=1
plt.barh(list(city_name),width= list(data_city.values()))
plt.title('Tầng suất đơn hàng theo city'.upper())
plt.show()
```

```python
# Vẽ pie chart
kq = plt.pie(list(data_city.values()),startangle=90,autopct="%1.1f%%",explode = [0,0,0.1],shadow=True)[0][0].set_alpha(0.35)
plt.legend(city_name,loc=[1,0.8])
plt.show()
```

```python
# Tạo biến chưa số liệu theo giới tính
data_gender = dict.fromkeys(gender_name,0)

for gender in data_gender:
    for row in data:
        if(gender == row['Gender']): data_gender[gender] +=1
plt.bar(list(gender_name),height= list(data_gender.values()))
plt.title('Tầng suất đơn hàng theo gender'.upper())
plt.show()
```

```python
kq = plt.pie(list(data_gender.values()),startangle=90,autopct="%1.1f%%",explode = [0,0.1],shadow=True)[0][0].set_alpha(0.35)
plt.legend(gender_name,loc=[1,0.8])
plt.show()
```

# Vẽ biểu đồ histogram|

```python
Gia_Tri_Don_Hang = []
rating = []
for row in data:
    Gia_Tri_Don_Hang.append(float(row['Total']))
    rating.append(float(row['Rating']))
```

```python
# Phân bố chung theo giá trị đơn hàng và rating
fig = plt.figure(figsize=(10,8))

ax1 = fig.add_subplot(1,2,1)
ax1.hist(Gia_Tri_Don_Hang,bins = 50, facecolor ='blue', alpha=0.55, edgecolor='k',label='Gia Tri Don Hang')
ax1.legend()

ax2 = fig.add_subplot(1,2,2)
ax2.hist(rating, bins=50,facecolor='red', alpha=0.55, edgecolor='k', label='rating')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
# Phân bố giá trị đơn hàng theo giới tính
data_gender_donhang = {'Nữ':[], 'Nam':[]}
data_gender_rating = {'Nữ':[], 'Nam':[]}

for gender in data_gender_donhang:
    for row in data:
        if(gender == row['Gender']): 
            data_gender_donhang[gender].append(row['Total'])
            data_gender_rating[gender].append(row['Rating'])
```

```python
# Phân bố trên cùng 1 biểu đồ cho đơn hàng
fig = plt.figure(figsize = (10,8))

#donhang
axes_dh = fig.add_subplot(1,2,1)
axes_dh.hist([float(i) for i in data_gender_donhang['Nữ']], bins=50, facecolor='red', density = False, alpha=0.75,edgecolor="k", label="Female")
axes_dh.hist([float(i) for i in data_gender_donhang['Nam']],bins=50,facecolor="blue",density=False,alpha=0.55,edgecolor="k",label="Male")
axes_dh.legend()
axes_dh.grid()
axes_dh.set_title('Gia Tri Don Hang')
#rating
axes_rating = fig.add_subplot(1,2,2)
axes_rating.hist([float(i) for i in data_gender_rating['Nữ']], bins=50, facecolor='red', density = False, alpha=0.75,edgecolor="k", label="Female")
axes_rating.hist([float(i) for i in data_gender_rating['Nam']],bins=50,facecolor="blue",density=False,alpha=0.55,edgecolor="k",label="Male")
axes_rating.legend()
axes_rating.grid()
axes_rating.set_title('Rating')
plt.show()
```

```python
#Phân bố trên 2 axes
fig = plt.figure(figsize = (10,8))

#Female
dh_ax1 = fig.add_subplot(2,2,1)
dh_ax1.hist([float(i) for i in data_gender_donhang['Nữ']], bins=50, facecolor='red', density = True, alpha=0.75,edgecolor="k", label="Female")
dh_ax1.legend()
dh_ax1.set_title('Female - Don Hang')
#Male
dh_ax2 = fig.add_subplot(2,2,2, sharey =dh_ax1 )
dh_ax2.hist([float(i) for i in data_gender_donhang['Nam']], bins=50, facecolor='red', density = True, alpha=0.75,edgecolor="k", label="Male")
dh_ax2.legend()
dh_ax2.set_title('Male - Don Hang')
dh_ax2.get_yaxis().set_visible(False)

#Female - Rating
rating_ax1 = fig.add_subplot(2,2,3)
rating_ax1.hist([float(i) for i in data_gender_rating['Nữ']], bins=50, facecolor='Blue', density = False, alpha=0.75,edgecolor="k", label="Female")
rating_ax1.legend()
rating_ax1.set_title('Female-rating')

#Male - Rating
rating_ax2 = fig.add_subplot(2,2,4, sharey = rating_ax1)
rating_ax2.hist([float(i) for i in data_gender_rating['Nam']], bins=50, facecolor='Blue', density = False, alpha=0.75,edgecolor="k", label="Male")
rating_ax2.legend()
rating_ax2.set_title('Male-rating')

plt.tight_layout()
plt.show()
```

# Vẽ biểu đồ tree-map

```python
# Lấy danh sách product_line
data_product_line = []
for row in data:
    data_product_line.append(row['Product line'])
product_line_name = list(set(data_product_line))

# Lấy giá trị tổng đơn hàng của từng product_line
total_product_line = {}
for product in product_line_name:
    total_product_line[product] = 0
    for row in data:
        if(product == row['Product line']):total_product_line[product] += float(row['Total'])
```
